```python
from datasets.download_data import download_data
from datasets.consts import Dataset, DatasetType
from datasets.get_data_loader import get_data_loader
import matplotlib.pyplot as plt
import torchvision
import numpy as np
```

```python
dataset = Dataset.CUB200
```

```python
transforms = [
    torchvision.transforms.Resize((224, 224)),
    # torchvision.transforms.Lambda(lambda x: x / 255.0)
    torchvision.transforms.ToTensor()
]
transforms = torchvision.transforms.Compose(transforms)

test = download_data(dataset, DatasetType.TEST, transform=transforms)
```

```python
data_loader = get_data_loader(test, 5, 0, 1, False)
data_iter = iter(data_loader)
```

```python
images, labels = next(data_iter)
```

```python
plt.imshow(np.transpose(images[0][0], (1, 2, 0)))
```

```python
def show_images(images) -> None:
    n: int = len(images)
    f = plt.figure(figsize=(20, 20//n))
    for i in range(n):
        f.add_subplot(1, n, i + 1)
        plt.imshow(np.transpose(images[i], (1, 2, 0)))
        
        plt.axis('off')
    plt.tight_layout()
    plt.savefig(f'../images/{dataset.name}_images.png')
    plt.show(block=True)
```

```python
show_images(images[0])
```
